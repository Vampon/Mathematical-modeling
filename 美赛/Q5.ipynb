```python
genre=pd.read_csv(r"genreCharctis.csv",header=0)
genre.drop(['artist_id','count'],axis=1,inplace=True)
genreList=genre['genre'].tolist()
```

```python
tem1=data.T
genre.set_index(['genre'],inplace=True)
tem2=genre.T
```

```python
ans=[]#利用维度距离方法
result=0
for i in range(1921,2021,1):
    z=[]
    x3=tem1[i].tolist()
    for j in range(0,20):
        x4=tem2[genreList[j]].tolist()
        for k in range(0,13):
            result=result+(x4[k]-x3[k])**2
        re=result**0.5
        result=0
        z.append(re)
    index=z.index(min(z))
    ans.append(genreList[index])
```

```python
data['genre']=ans
```

```python
data.to_csv("year_with_genre_distance.csv")
```

```python
def cos_sim(vector_a, vector_b):
    """
    计算两个向量之间的余弦相似度
    :param vector_a: 向量 a 
    :param vector_b: 向量 b
    :return: sim
    """
    vector_a = np.mat(vector_a)
    vector_b = np.mat(vector_b)
    num = float(vector_a * vector_b.T)
    denom = np.linalg.norm(vector_a) * np.linalg.norm(vector_b)
    cos = num / denom
    sim = 0.5 + 0.5 * cos
    return sim
```

```python
ans=[]#利用余弦相似度方法
result=0
for i in range(1921,2021,1):
    z=[]
    x3=tem1[i].tolist()
    for j in range(0,20):
        x4=tem2[genreList[j]].tolist()
        re=cos_sim(x3,x4)
        z.append(re)
    index=z.index(max(z))
    ans.append(genreList[index])
```
